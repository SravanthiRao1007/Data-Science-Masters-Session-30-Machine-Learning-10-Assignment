```python
## Data Science Masters Session 30 Machine Learning 10 Assignment Sravanthi
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
# Randomly seed with some numpy array
np.random.seed(5)
```

```python
# centers
centers = [[1, 1], [-1, -1], [1, -1]]
```

```python
# Load the dataset and assign x and y
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# Clear current fig and create axes 3d object
fig = plt.figure(1, figsize=(4, 3))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

# clear the axis and fit the model with the data available
plt.cla()
pca = decomposition.PCA(n_components=3)
pca.fit(X)

# Perform a dimensinality reduction
X = pca.transform(X)

# Create labels
for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(X[y == label, 0].mean(),
              X[y == label, 1].mean() + 1.5,
              X[y == label, 2].mean(), name,
              bbox=dict(edgecolor='w', facecolor='w'))
    
# align labels and plot the graph
y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, edgecolor='k')
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
plt.show()
```
